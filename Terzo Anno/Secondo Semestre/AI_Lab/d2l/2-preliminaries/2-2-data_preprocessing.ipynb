# Data Preprocessing
Per utilizzare il deep learning avremo bisogno di estrarre grandi quantità di dati ed adattarli alle nostre esigenze. Nella maggior parte dei casi utilizzeremo la libreria **Pandas**.

Facciamo un esempio con un file CSV dove ogni riga indica appunto una riga e le colonne sono separate dalle virgole. Scriviamo un file e poi lo carichiamo utilizzando Pandas

```python
import os

os.makedirs(os.path.join('..', '..', 'data'), exist_ok=True)
data_file = os.path.join('..', '..', 'data', 'house_tiny.csv')
with open(data_file, 'w') as f:
    f.write('''NumRooms,RoofType,Price
NA,NA,127500
2,NA,106000
4,Slate,178100
NA,NA,140000''')
```

```python
import pandas as pd

data = pd.read_csv(data_file)
print(data)
```

## Data Preparation
Possiamo selezionare le colonne o tramite i nomi oppure tramite un indice intero, questo meccanismo si chiama integer-location based indexing (iloc).

Notiamo che i dati mancanti sono stati rimpiazzati da Pandas con un valore speciale _NaN_ (Not a Number).
Questi valori mancanti possono essere risolti in 2 modi, o scartando le righe che hanno dati mancanti oppure facendo inferenza su questi ovvero cercare di predirli in base hai dati che abbiamo.

Ovviamente ci sono diversi modi per fare inferenza, ad esempio nella tabella precedente notiamo che `RoofType` può assumere il valore `Slate` o `NaN`, Pandas può convertire questa colonna in due colonne `RoofType_Slate` e `RoofType_nan` con valori booleani per indicare se una riga appartiene a Slate o NaN.

```python
# Come input abbiamo le prime due colonne mentre come target il prezzo (l'ultima)
inputs, targets = data.iloc[:, 0:2], data.iloc[:, 2]
inputs = pd.get_dummies(inputs, dummy_na=True)
print(inputs)
```

Un altro metodo per sotituire i valori NaN è quello di effettuare la media degli altri valori.

```python
inputs = inputs.fillna(inputs.mean())
print(inputs)
```

## Conversion to the Tensor Format
Adesso che abbiamo gli inputs ed i target in formato numerico possiamo caricarli sottoforma di tensors.

```python
import torch

X = torch.tensor(inputs.to_numpy(dtype=float))
y = torch.tensor(targets.to_numpy(dtype=float))
X, y
```
