La struttura base che utilizziamo per memorizzare e manipolare i dati sono array multidimensionali che chiamiamo tensors.

```python
import torch
```

PyTorch fornisce molte funzioni utili per creare dei tensor con dei dati già all'interno. Ad esempio con `arange(n)` creiamo un tensor con numeri che partono da 0 incluso fino ad n escluso con un intervallo che possiamo specificare, di default è 1. Se non viene specificato il tensor viene creato in RAM e potrà essere utilizzato soltanto dalla CPU.

```python
x = torch.arange(12, dtype=torch.float32)
x
```

Ogni valore viene chiamato _element_ del tensor. Il tensor x contiene 12 elementi. Possiamo vedere quanti elementi ha un tensor con

```python
x.numel()
```

I tensor sono multidimensionali e avere quindi più assi, per scoprire la lunghezza su ogni asse e vedere quindi la shape di un tensor chiamiamo appunto `shape`

```python
x.shape
```

Possiamo cambiare la shape di un tensor senza alterarne i valori o la grandezza chiamano `reshape`, ad esempio possiamo trasformare il tensor di lunghezza 12 in una matrice 3x4.
Il nuovo tensor manterrà tutti gli elementi ma li riorganizzerà in una matrice.

```python
X = x.reshape(3, 4)
X
```

Specificare tutte le grandezze nel metodo `reshape` è ridondante, infatti siccome già sappiamo la grandezza del tensor possiamo ricavare alcune dimensioni a partire da altre. Ad esempio dato un tensor di grandezza $n$ che vogliamo trasformare in $(h,w)$ sappiamo che $w = n/h$. Per fare questo ci basta inserire $-1$ nel campo che vogliamo far calcolare in automatico, quindi nel codice precedente avremmo potuto scrivere `x.reshape(3,-1)` oppure `x.reshape(-1, 4)`.

PyTorch mette a disposizione delle funzioni per creare tensor con tutti 1 o 0 e specificare anche le dimensioni di ogni asse

```python
torch.zeros((2,3,4))
```

```python
torch.ones((2,3,4))
```

Possiamo anche creare tensor con valori casuali, utile ad esempio per inizializzare i parametri di una rete neurale.
Con il seguente blocco creiamo un tensor con elementi scelti a caso da una distribuzione Gaussiana

```python
torch.randn(3,4)
```

Ovviamente possiamo anche fornire noi i dati per ogni elemento attraverso liste Python

```python
torch.tensor([[2, 1, 4, 3], [1, 2, 3, 4], [4, 3, 2, 1]])
```

## Indexing and Slicing
Possiamo accedere agli elementi dei tensors in modo simile alle liste Python, possiamo utilizzare anche lo slicing per accedere a range di valori, quindi usare `X[start:stop]` per ottenere tutti i valori da `start` incluso a `stop` escluso.
Se in un array multidimensionali utilizziamo un solo indice allora questo farà riferimento all'asse 0.

```python
X[-1], X[1:3]
```

Possiamo le posizioni anche per scrivere dati

```python
X[1, 2] = 17
X
```

O anche lo slicing, in questo caso sovrascriviamo tutte le righe fino alla 2 esclusa, e di queste tutte le colonne.

```python
X[:2, :] = 12
X
```

## Operations
Adesso possiamo iniziare a fare operazioni con i tensors, tra le più utili ci sono quelle element-wise ovvero che applicano operazioni scalari su ogni elemento del tensor. Per operazioni che prendono in input due tensor, questo tipo di operazioni lavora su coppie di elementi.

Usando una notazione matematica indichiamo queste operazioni unarie (1 input) con $f:\mathbb{R} \to \mathbb{R}$.

Molte operazioni standard possono essere applicate element-wise, ad esempio $e^x$

```python
torch.exp(x)
```

Possiamo invece indicare le funzioni binarie con $f:\mathbb{R},\mathbb{R}\to\mathbb{R}$. Presi due vettori $u$ e $v$ della stessa shape e una funzione binaria $f$ possiamo creare un nuovo vettore $c=F(u,V)$ impostando $c_i = f(u_i, v_i)$ per ogni $i$, dove $c_i,u_i,v_i$ sono l'$i$-esimo elemento dei vettori $c,u,v$.

```python
x = torch.tensor([1.0, 2, 4, 8])
y = torch.tensor([2, 2, 2, 2])
x + y, x - y, x * y, x / y, x ** y
```

È anche possibile effettuare operazioni di algebra lineare come prodotto tra matrici ma le vedremo più avanti.

Si possono concatenare più tensors fra loro, basta fornire in input quest'ultimi e anche su che asse effettuare la concatenazione

```python
X = torch.arange(12, dtype=torch.float32).reshape((3,4))
Y = torch.tensor([[2.0, 1, 4, 3], [1, 2, 3, 4], [4, 3, 2, 1]])
torch.cat((X, Y), dim=0), torch.cat((X, Y), dim=1)
```

Nel primo caso stiamo concatenando le righe e infatti otteniamo 6 come lunghezza dell'asse 0 ovvero la somma dello stesso asse per i due tensors. Nel secondo output invece abbiamo concatenato sulle colonne e infatti l'asse 1 ha come valore 8 ovvero la somma dello stesso asse per i due tensors.

È possibile costruire tensors anche a partire da operatori logici, ad esempio con X == Y, in questo caso avremo un tensor che per ogni posizione $i,j$ se $X[i,j] == Y[i,j]$ allora avrà come valore 1 altrimenti 0.

```python
X == Y
```

Ci sono anche operazioni che convertono tensor in altri tensor composti da singoli elemnti

```python
X.sum()
```

## Broadcasting
Le operazioni viste fino ad adesso possiamo eseguirle soltanto su tensors con la stessa dimensione. In alcuni casi però anche con shape differenti possiamo comunque eseguire operazioni elementwise, dobbiamo utilizzare il meccanismo di broadcassting.
Il broadcasting si applica in due step:
- Espandere uno o entrambi gli array sull'asse che ha lunghezza 1 copiando gli elementi
- Effettuare l'operazione su questi nuovi array

```python
a = torch.arange(3).reshape((3,1))
b = torch.arange(2).reshape((1,2))
a,b
```

In questo caso abbiamo $a$ e $b$ con, rispettivamente, shape 3x1 e 1x2, con il broadcasting otteniamo una matrice 3x2. La prima matrice a viene replicata sulle colonne, quindi diventerà una matrice 3x2 copiano la prima colonna sulla seconda. La matrice b diventerà anch'essa 3x2 ma copieremo la prima riga sulle altre.
Una volta effettuate queste operazione viene effettuata la somma elementwise su queste due nuove matrici, infatti otteniamo

```python
a + b
```

## Saving Memory
Eseguire operazioni comporta l'utilizzo di altra memoria nell'host, ad esempio se scriviamo Y = X + Y stiamo deferenziando il puntatore al tensor Y per fargli puntare una nuova zona di memoria, possiamo verificarlo in Python

```python
before = id(Y)
Y = Y + X
id(Y) == before
```

Questo comportamento potrebbe essere scomodo per due motivi principali
- Non vogliamo allocare memoria inutilmente, in machine learning spesso avremo megabytes di parametri da aggiornare molte volte al secondo, se possibile vogliamo effettuare queste operazioni _in place_.
- In alcuni casi abbiamo diverse variabili che puntano allo stesso parametro, se non lo aggiorniamo in place dobbiamo prestare attenzione ad aggiornare tutte le reference a quel paraemtro.

Effettuare operazioni in place è semplice. Per farlo ci basta assegnare l'operazione ad un array precedente utilizzando la notazione dello slicing `Y[:]=expression`

```python
Z = torch.zeros_like(Y)
print(f"id(Z): {id(Z)}")
Z[:] = X + Y
print(f"id(Z): {id(Z)}")
```

In realtà in questo caso se X non viene riutilizzato più avanti possiamo allocare tutto al suo interno nello stesso modo

```python
before = id(X)
X += Y # Oppure X[:] = X + Y
id(X) == before
```
