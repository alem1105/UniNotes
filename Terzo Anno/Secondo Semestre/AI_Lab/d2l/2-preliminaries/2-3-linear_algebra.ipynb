## Scalars
Indichiamo dei semplici numeri con il termine **scalari**. Ad esempio se in un luogo abbiamo una temperatura di 72 gradi Fahrenheit e vogliamo convertirla in gradi Celsius dobbiamo calcolare l'espressione $c=\frac{5}{9}(f-32)$ impostando $f=72$. In questa equazione i valori $5,9,32$ sono scalari costanti mentre le variabili $c,f$ sono sempre scalari ma per ora sconosciuti.

Indichiamo gli scalari con lettere lower-case ($x,y,z$) e lo spazio dei valori reali di tutti gli scalari con $\mathbb{R}$. Con l'espressione $x \in \mathbb{R}$ stiamo dicendo che $x$ è uno scalare con valore reale.

Gli scalari sono implementati come tensors che contengono un solo elemento

```python
x = torch.tensor(3.0)
y = torch.tensor(2.0)

x+y, x*y, x/y, x**y
```

## Vectors
Possiamo immaginare un vettore come un array di scalari con lunghezza fissa. Chiameremo questi scalari _elementi_ del vettore.
Quando trattiamo dati reali, i valori dei vettori avranno un significato reale, ad esempio se stiamo allenando un modello per studiare il rischio di attacchi di cuore ogni vettore rappresenterà un paziente e ogni elemento corrisponderà ad un parametro come il livello di colesterolo, i minuti di esercizio al giorno ecc...
Indichiamo i vettori con lettere lower-case in grassetto (**x,y,z**) e i loro elementi con un indice scalare in basso a destra ($\pmb{x}_2$)

I vettori sono implementati come tensors ad una dimensione.

```python
x = torch.arange(3)
x
```

Per indicare che un vettore contiene $n$ elementi scriviamo $\pmb{x}\in\mathbb{R}^n$, chiamiamo $n$ la dimensione del vettore. Nel codice questa corrisponde alla length che è accessibile tramite le funzioni built-in di Python

```python
len(x)
```

Possiamo accedere alla lunghezza anche con l'attributo shape dei tensor, la shape è una tupla che indica la lunghezza del tensor su ogni asse, in questo caso uno solo.

```python
x.shape
```

Per non fare confusione con i termini, spesso utilizzati per indicare entrambe le cose:
- Order: Numero di assi
- Dimensionality: Numero di elementi su un asse

## Matrices
Le matrici sono tensor del secondo ordine e le indichiamo con lettere upper-case e in grassetto ($\pmb{X,Y,Z}$). Con l'espressione $\pmb{A}\in\mathbb{R}^{m \times n}$ indichiamo che la matrice $\pmb{A}$ contiene $m \times n$ scalari reali organizzati in $m$ righe e $n$ colonne.
Quando $m=n$ diciamo che la matrice è quadrata.
Per indicare un elemento lo facciamo con una lettere lowercase, di solito quella della matrice e i due indici in basso a sinistra, quindi $a_{ij}$ è l'elemento di $\pmb{A}$ alla riga $i$ e alla colonna $j$.

```python
A = torch.arange(6).reshape(3,2)
A
```

Possiamo invertire gli assi delle matrici ovvero scambiare le righe con le colonne, il risultato di questa operazione è la trasposta di una matrice.
Indichiamo la trasposta di $\pmb{A}$ con $\pmb{A}^T$ e se $\pmb{B}=\pmb{A}^T$ allora $b_{ij}=a_{ji}$.
Notiamo quindi che la trasposta di una matrice $m\times n$ è una matrice $n\times m$

In PyTorch effettuiamo questa operazione con:

```python
A.T
```

Una matrice si dice **simmetrica** quando è uguale alla sua trasposta. La seguente matrice è simmetrica

```python
A = torch.tensor([[1,2,3],[2,0,4],[3,4,5]])
A == A.T
```

Le matrici sono estremamente utili per rappresentare i datasets, di solito le righe indicano dei singoli records mentre le colonne i loro attributi.

## Basic Properties of Tensor Arithmetic
Una delle principali proprietà è che i risultati di operazioni elementwise mantengono la shape degli operandi.

```python
A = torch.arange(6, dtype=torch.float32).reshape(2,3)
B = A.clone() # Alloca nuova memoria
A, A + B
```

Il prodotto elementwise di due matrici si chiama _Hadamart product_ e lo indichiamo con $\odot$

```python
A * B
```

Sommare o moltiplcare uno scalare ed un tensor produce un risultato con la stessa shape del tensor originale

```python
a = 2
X = torch.arange(24).reshape(2,3,4)
a + X, (a * X).shape
```

## Reduction
Possiamo calcolare la somma di tutti gli elementi di un tensor con la funzione sum.

```python
x = torch.arange(3, dtype=torch.float32)
x, x.sum()
```

Questo tipo di operazione si muove in tutte le assi, vediamo ad esempio come si comporta con una matrice

```python
A.shape, A.sum()
```

Quindi di default chiamare l'operazione sum agisce su tutti gli assi, possiamo però specificare su quali effettuare l'operazione, fare questo comporta la rimozione dell'asse specificato dal risultato

```python
A.shape, A.sum(axis=0).shape
```

Ovviamente lo stesso accade per le colonne

```python
A.shape, A.sum(axis=1).shape
```

Possiamo anche calcolare la media con la funzione mean

```python
A.mean(), A.sum() / A.numel()
```

Ovviamente valgono le stesse regole di prima sulla specifica degli assi.

## Non-Reduction Sum
In alcuni casi potrebbe essere utile mantenere invariato il numero di assi quando invochiamo la funzione per calcolare la somma o la media, ad esempio quando sappiamo che utilizzeremo il meccanismo di broadcast.

```python
print(A)
sum_A = A.sum(axis=1, keepdim=True)
sum_A, sum_A.shape
```

Quindi, quando noi specifichiamo un asse stiamo dicendo a PyTorch di muoversi lungo quell'asse e sommare gli elementi per poi "distruggere" quell'asse. Se specifichiamo keepdim l'asse non verrà distrutto ma verrà ridotto a grandezza 1.

Dato che in questo modo manteniamo la shape possiamo quindi effettuare operazioni successive con il tensor dato che poi verrà applicato il broadcasting, adesso dividere ogni elemento di una riga per la somma della riga e ottenere righe che quindi hanno come somma 1.

```python
A / sum_A
```

Se vogliamo calcolare la somma cumulativa degli elementi su un certo asse senza ridurre la shape possiamo usare la funzione `cumsum`

```python
A.cumsum(axis=0)
```

## Dot Products
Una delle principali operazioni in algebra lineare è il dot product, dati due vettori $\pmb{x,y}\in\mathbb{R}^d$ il loro dot product $\pmb{x}^T\pmb{y}=\sum_{i=1}^d x_iy_i$. Anche conosciuto come _inner product_ $<\pmb{x,y}>$

```python
y = torch.ones(3, dtype=torch.float32)
x, y, torch.dot(x, y)
```

Questo può essere ottenuto anche effettuando prima una moltiplicazione elementwise e poi una somma.

```python
torch.sum(x * y)
```

Il dot products è utile in molte occasioni, ad esempio dato un set di valori in un vettore $\pmb{x}\in\mathbb{R}^n$ e un set di pesi nel vettore $\pmb{w}\in\mathbb{R}^n$, la somma ponderata dei valori può essere espressa come il dot product $\pmb{x}^T\pmb{w}$. Quando i pesi sono non negativi e la loro somma è 1 allora il dot product rappresenta una media ponderata.

## Matrix-Vector Products
Possiamo vedere come calcolare il prodotto tra una matrice $m\times n \ \pmb{A}$ e un vettore a $n$ dimensioni $\pmb{x}$.

Per semplificare le cose immaginiamo la matrice $\pmb{A}$ come un vettore con $m$ righe composta da vettori lunghi $n$, chiamiamo questi vettori $a_1^T,a_2^T,...,a_m^T$. Adesso prendiamo ogni riga ed effettuiamo il prodotto con il vettore $\pmb{x}$, il risultato sarà un elemento del nuovo vettore.

Nel codice utilizziamo la funzione `mv` per effettuare questo prodotto, ovviamente per poter funzionare la dimensione delle colonne di $\pmb{A}$ deve essere la stessa lunghezza di $\pmb{x}$. Python ci offre anche un altro operatore che permette di effettuare sia prodotti matrici-vettori che matrici-matrici ovvero `@`.

```python
print(A)
print(x)
A.shape, x.shape, torch.mv(A,x), A@x
```

## Matrix-Matrix Multiplication
Immaginiamo di avere due matrici $\pmb{A}\in\mathbb{R}^{n\times k}$ e $\pmb{B}\in\mathbb{R}^{k\times m}$.

Adesso indichiamo con $\pmb{a}_i^T\in\mathbb{R}^k$ il vettore (riga) che rappresenta la $i$-esima riga della matrice $\pmb{A}$, mentre con $\pmb{b}_j\in\mathbb{R}^k$ il vettore (colonna) che rappresenta la colonna $j$-esima della matrice $\pmb{B}$.

Per formare la matrice risultato $\pmb{C}\in\mathbb{R}^{n\times m}$ dobbiamo calcolare ogni elemento $c_{i j}$ come il dot product tra la $i$-esima riga di $\pmb{A}$ e la $j$-esima colonna di $\pmb{B}$

```python
B = torch.ones(3, 4)
print(A)
torch.mm(A, B), A@B
```

## Norms
Un'altra categoria di operazioni utilissime in algebra sono le _norms_. In generale la norm di un vettore ci dice "quanto è grande" ovvero qual è la sua distanza dall'origine, ad esempio la $l_2$ norm misura la lugnezza Euclidea di un vettore.

Una norm function $||\cdot||$ è una funzione che mappa un vettore in uno scalare e soddisfa le 3 proprietà seguenti:
- Dato un vettore $\pmb{x}$ e uno scalare, se li moltiplichiamo allora la sua lunghezza si moltiplica per quel numero. $$||\alpha \pmb{x}|| = |\alpha|||\pmb{x}||$$
- Presi due qualsiasi vettori $\pmb{x,y}$ è soddisfatta la **Disuguaglianza Triangolare*, ovvero una linea retta tra i due punti è sempre il percorso minore $$||\pmb{x}+\pmb{y}||\leq ||\pmb{x}||+||\pmb{y}||$$
- Le lunghezze non possono essere negative, l'unico modo per avere lunghezza zero è che il vettore non si è mai mosso dall'origine. $$||\pmb{x} > 0|| \ \ \forall \pmb{x}\neq 0$$

La $l_2$ norm equivale al calcolo del teorema di Pitagora infatti è data da: $$||\pmb{x}||_2=\sqrt{\sum_{i=1}^n x_i^2}$$

```python
u = torch.tensor([3.0, -4.0])
torch.norm(u)
```

Che infatti equivale a $\sqrt{3^2 + (-4)^2}=\sqrt{9 + 16}=5$

La $l_1$ norm invece è associata alla Manhattan distance, per definizione questa somma tutti i valori assoluti degli elementi e quindi abbiamo che: $$||\pmb{x}||_1=\sum_{i=1}^n |x_i|$$

```python
torch.abs(u).sum()
```

Entrambe $l_1, l_2$ sono un caso speciale della norm generale $l_p$: $$||\pmb{x}||_p = \left(\sum_{i=1}^n |x_i|^p\right)^\frac{1}{p}$$

Estendere questo concetto alle matrici diventa però più complicato, come misuriamo "quanto sono grandi"? Introduciamo la **Norma di Frobenius**, possiamo vederla come il teorema di Pitagora applicato ad una matrice. $$|| \pmb{X} ||_F=\sqrt{\sum_{i=1}^m \sum_{j=1}^n x^2_{i j}}$$

```python
torch.norm(torch.ones(4, 9))
```

Queste norm ci servono per misurare appunto la distanza tra i vettori dei risultati che ci aspettiamo e quelli predetti dal nostro modello. Addestrare una rete significa appunto cercare di rendere questa norma il più vicino possibile allo 0.
